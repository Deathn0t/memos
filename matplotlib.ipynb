# Scatter plot wiht heatmap

```python

from matplotlib import pyplot as plt
from matplotlib import cm as cm
import numpy as np

size = 100
x = np.random.uniform(0, 1, size)
y = np.random.uniform(0, 1, size)
z = np.random.uniform(0, 1, size)

marker_size = 400
plt.scatter(x,y,
            edgecolors='none',
            s=marker_size,
            c=z,
            marker="^",
            cmap=cm.autumn_r,
            alpha=1)

cb = plt.colorbar()

plt.show()
```
